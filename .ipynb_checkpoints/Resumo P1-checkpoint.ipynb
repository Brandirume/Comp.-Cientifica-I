# Bisseção com tolerancia:
Método simples de calculo da raiz por bisseção, escolhe um intervalo [a,b] e se f(a) vezes f(b) é menor q zero então a função passa por zero, logo existe uma raiz entre esse intervalo.
O metodo da bisseção usa isso para tentar se aproximar ao maximo da raiz da função, sempre cortando o 
intervalo ao meio.

```python
def bissecao_step(f, a, b):
    z = (a+b)/2
    if f(a)*f(z)<0:
        b=z
    else:
        a=z
    return a,b
def bissecao(f, a, b, tol=1e-8, err=1e-8):
    e=abs(b-a)
    j=0
    i=0
    while True:
        e=abs(b-a)
        if e<tol or abs(f((a+b)/2))<err:
            break
        j +=3
        i +=1
        a,b = bissecao_step(f, a, b)
        e=abs(b-a)
    return (a+b)/2,i,j
```

# Bisseção com Erro para graficos:

```python
def Err_biss(a,b):
    r=(a+b)/2
    E_absoluto = abs((b-a)/2)
    if a*b<=0:
        return E_absoluto, +infty
    if a<b:
        E_relativo = abs((r-b)/b)
    else:
        E_relativo = abs((r-a)/a)
    return E_absoluto, E_relativo
def bissecao_step(f,a,b):
    z = (a+b)/2
    if f(a)*f(z)<0:
        b=z
    else:
        a=z
    return a,b
def bissecao_extremidades(f,a,b, abstol=1e-10,reltol=1e-10):
    """ Função que retorna as uma lista contendo os intervalos que a função f possui uma raiz, o metodo 
    usado é igual ao da função bissecao"""
    l=[]
    while True:
        l.append([a,b])
        e_abs, e_rel = Err_biss(a,b)
        if e_abs<abstol and e_rel<reltol:
            break
        a,b = bissecao_step(f, a, b) 
    return l
```

# Bisseção para graficos sem Erro:

```python
def bissecao_step(f,a,b):
    z = (a+b)/2
    if f(a)*f(z)<0:
        b=z
    else:
        a=z
    return a,b
def bissecao_pts(f, a, b, xtol=1e-8, ytol=1e-8):
    """
    Encontra uma raiz de  f  pelo algoritmo da bissecao, retornando
    todos os pontos pelos quais o algoritmo passa.
    
    O algoritmo termina quando o tamanho do intervalo ficar menor do que  xtol,
    ou quando o valor absoluto da função ficar abaixo de  ytol.
    """
    pts=[a, b]
    while True:
        z=(a+b)/2
        pts.append(z)
        if abs(f(z))<ytol or abs(b-a)<xtol:
            break
        a,b = bissecao_step(f, a, b)
    return pts
```

# Newton:

```python
def newton_pts(f, df, x, xtol=1e-8, ytol=1e-8, maxiter=100):
    l=[x]
    for i in range(maxiter):
        passo = f(x)/df(x)
        newx= x-passo
        l.append(newx)
        if abs(passo) < xtol or abs(f(newx)) < ytol:
            return l
        x=newx
    return l
```

# Comparação entre o metodo da bisseção e de newton:

```python
def F(x):
    return -2*x**5+x**4+2*x**2+3*x+4
def f(x):
    return -10*x**4+4*x**3+4*x-3
raiz=1.5872110603237
bi=bissecao_pts(F,1.5,1.7,1e-12,1e-12)
err1=[]
newton=newton_pts(F,f,1.7,1e-12,1e-12)
err2=[]
for x in bi:
    err1.append(abs(raiz-x))
for x in newton:
    err2.append(abs(raiz-x))
plt.semilogy(err1,label='bisseção')
plt.semilogy(err2,label='newton')
plt.legend()
```

# Derivadas numericas:

```python
def df_central(f, x, h=1e-5):
    """Derivada central de  f  em  x, com passo  h"""
    return (f(x+h) - f(x-h))/(2*h)
def ddf_central(f,x, h=1e-4):
    """Aproximação da segunda derivada de  f  em  x, usando a derivada central iterada com passo  h."""
    return (df_central(f,x+h)-df_central(f,x-h))/(2*h)
```

# Precisão da derivada central

```python
def G(x):
    return x**3-4*x**2+2*x+3
def g(x):
    return 3*x**2-8*x+2
steps=np.linspace(0,2,1000)
err=[]
for x in steps:
    err.append(abs(g(x)-df_central(G,x,1e-5)))
plt.semilogy(err)
```

# Método de Euler:
Usado pra calcular uma aproximação da função através de sua derivada.
é como se você usasse as retas que o metodo de newton aproxima as funções mas no caminho contrario

```python
def euler(F, t0,tf, f0, nsteps=100):
    """F é a equação diferencial
    to o ponto inicial
    tf o tempo final
    e f0 a condição inicial"""
    l=list(np.linspace(t0,tf,nsteps+1))
    f=[f0]
    for i in range(len(l)-1):
        fn=f[i]+F(l[i],f[i])*(l[i+1]-l[i])
        f.append(fn)
    return f
```

# Calculando o Erro do metodo de euler pra função dele:
Se alterar a função para um intervalo diferente de [0,1] o erro cresce de forma exponencial

```python
def E(x):
    return np.exp(x)
def e(x,y):
    return y
preciso=[E(x) for x in np.linspace(0,1,10000)]
aprox = euler(e, 0, 1, 1,10000)
Erro=[]
i=0
while i<len(preciso):
    Erro.append(abs(preciso[i]-aprox[i]))
    i+=1
fig,(x1,x2)=plt.subplots(1,2)
x1.plot(preciso)
x1.plot(aprox)
x2.loglog(Erro)
x1.set_title("grafico das funções")
x2.set_title("erro")
```

# método do retangulo:
Basicamente calcula a integral de uma função atraves de retangulos.
Sem muito a comentar pois vemos isso em calculo 1.
O erro de aproximações pra funções é claramente muito grande dependendo do intervalo
<img src='retangulo.png'>

```python
def retangulo(f,a,b,n=100):
    xs, h = np.linspace(a, b, num=n, endpoint=False, retstep = True)
    return np.sum(f(xs))*h
```

# Erro do metodo do retangulo para $f(x)=3.x^2-8.x+2$

```python
ns=np.logspace(1,4,dtype=int)
ans=G(2)-G(0)
retang=np.array([retangulo(g,0,2,n)for n in ns])
plt.loglog(ns,abs(ans-retang))
```

# método do ponto médio:
usa retangulos tbm, mas pega o ponto medio da distancia do intervalo n e usa isso pra definir a altura do retangulo:
<img src='midpoint.png'>

```python
def midpoint(f,a,b,n=100):
    l, h = np.linspace(a,b, num=n, endpoint=False, retstep=True)
    if n == 1:
        h = abs(b-a)
    mids = l + h/2
    return np.sum(f(mids)*h)
```

# Erro do metodo do ponto medio para $f(x)=3.x^2-8.x+2$

```python
mid=np.array([midpoint(g,0,2,n)for n in ns])
plt.loglog(ns,abs(ans-mid))
```

# método de simpson:
Pelo o que entendi, ele usa uma aproximação de um polinomio de segundo grau de cada intervalo e então integra cada um desses pequenos intervalos:
<img src='simpson.png'>

```python
def simpson(f,a,b,n=100):
    l, h = np.linspace(a, b, num=n, endpoint=False, retstep=True)
    if n == 1:
        h = abs(b-a)
    mids = l + h/2
    return (4*np.sum(f(mids)) + 2*np.sum(f(l)) + (f(b)-f(a))) * h/6
```

# Erro do metodo de simpson para $f(x)=3.x^2-8.x+2$

```python
simp=np.array([simpson(g,0,2,n)for n in ns])
plt.loglog(ns,abs(ans-simp))
```
